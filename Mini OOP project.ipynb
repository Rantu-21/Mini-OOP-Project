```python
class Burger_House:
    def __init__(self, dic):
        self.menu = dic
        self.shoplist = {}
               
    def display_Menu(self):
        print("We have the following burgers in our burger house")
        for i in self.menu:
            print(i, "----", self.menu[i])
            
    def add_to_shoplist(self, burg_nam):
        if burg_nam in self.menu:
            self.shoplist.update({burg_nam: self.menu[burg_nam]})
            print(f"{burg_nam} has been added to the shoplist")
        else:
            print(f"Sorry, we do not have {burg_nam} in our menu")
        
    def remove_from_shoplist(self, burg_nam):
        if burg_nam in self.shoplist:
            self.shoplist.pop(burg_nam)
            print(f"{burg_nam} has been removed from your shoplist")
        else:
            print(f"{burg_nam} is not in your shoplist")
        
    def confirm_order(self, name, address, ph_number):
        print(f"Thanks {name} for ordering from our burger house")
        print(f"Customer name : {name}")
        print(f"Customer's home address : {address}")
        print(f"Customer's contact number : {ph_number}")
        print("---------------------------------------------------------------------------------")
        total = 0
        for i in self.shoplist:
            total += self.shoplist[i]
            print(i, "                                     =", self.shoplist[i])
        print("---------------------------------------------------------------------------------")
        print("Total bill                                  =", total)
        
if __name__ == "__main__":
    rontu = Burger_House({"Chicken Burger": 120, "Beef Burger": 130, "Cheese Burger": 135, "BBQ Burger": 150})
    
    while True:
        print("Welcome to the Rantu Burger House. Enter your choice to continue")
        print("1. Display Menu")
        print("2. Add a Burger to shoplist")
        print("3. Remove a Burger from shoplist")
        print("4. Confirm your order")
        user_choice = input()
        if user_choice not in ["1", "2", "3", "4"]:
            print("Please enter a valid option")
            continue
        else:
            user_choice = int(user_choice)
        
        if user_choice == 1:
            rontu.display_Menu()
            
        elif user_choice == 2:
            burg_nam = input("Enter the name of burger you want to order: ")
            rontu.add_to_shoplist(burg_nam)
        
        elif user_choice == 3:
            burg_nam = input("Enter the name of burger you want to remove from your shoplist: ")
            rontu.remove_from_shoplist(burg_nam)
            
        elif user_choice == 4:
            name = input("Enter your name: ")
            address = input("Enter your address: ")
            ph_num = input("Enter your phone number: ")
            print("Congratulations!!! Your order has been successfully received")
            rontu.confirm_order(name, address, ph_num)
        
        print("Press q to quit or c to continue")
        user_choice2 = input()
        if user_choice2 == "q":
            break
        elif user_choice2 == "c":
            continue
        else:
            print("Invalid option, exiting the program.")
            break
```
